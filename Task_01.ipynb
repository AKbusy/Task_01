```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('titanic.csv')
data.head()
```

```python
#Age = data.Age
Age = data["Age"]
```

```python
plt.figure(figsize=(10,5))
plt.hist(Age, bins=50)
plt.title("Titanic Age Distribution",fontsize=16)
plt.xlabel("Age",fontsize=14)
plt.ylabel("Counts",fontsize=14)
plt.grid(axis="y")
plt.show()
```

```python
#Sex= data.Sex
Sex = data["Sex"]
```

```python
plt.figure(figsize=(10,5))
plt.hist(Sex, bins=7)
plt.title("Male and Female survived",fontsize=16)
plt.xlabel("Gender",fontsize=14)
plt.ylabel("Count",fontsize=14)
plt.grid(axis="y")
plt.show()
```
